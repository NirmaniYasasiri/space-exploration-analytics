```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/space_cleaned.csv')
print(df.shape)
df.head()
```

```python
df['mission_status'].value_counts().plot(kind='bar', color='steelblue')
plt.title('Mission Outcome Distribution')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../outputs/01_mission_outcomes.png')
plt.show()
```

```python
yearly = df.groupby('year').agg(
    launches=('is_success', 'count'),
    success_rate=('is_success', 'mean')
).reset_index()

fig, ax1 = plt.subplots(figsize=(14, 5))
ax1.bar(yearly['year'], yearly['launches'], color='steelblue', alpha=0.6, label='Launches')
ax2 = ax1.twinx()
ax2.plot(yearly['year'], yearly['success_rate'] * 100, color='red', linewidth=2, label='Success Rate %')
ax1.set_xlabel('Year')
ax1.set_ylabel('Launches')
ax2.set_ylabel('Success Rate %')
plt.title('Annual Launches & Success Rate (1957–2020)')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.tight_layout()
plt.savefig('../outputs/02_annual_trend.png')
plt.show()
```

```python
top_companies = df.groupby('company').agg(
    launches=('is_success', 'count'),
    success_rate=('is_success', 'mean')
).sort_values('launches', ascending=False).head(10).reset_index()

fig, ax1 = plt.subplots(figsize=(12, 5))
bars = ax1.barh(top_companies['company'], top_companies['launches'], color='steelblue')
ax2 = ax1.twiny()
ax2.plot(top_companies['success_rate'] * 100, top_companies['company'], 'ro', markersize=8)
ax1.set_xlabel('Total Launches')
ax2.set_xlabel('Success Rate %')
plt.title('Top 10 Companies by Launch Volume')
plt.tight_layout()
plt.savefig('../outputs/03_top_companies.png')
plt.show()
```
